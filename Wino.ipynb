```python

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
RANDOM_STATE=9
data = load_wine()
print(data.DESCR)

```

Podział zbiorów na tesowe oraz validacyjne. Wykorzystanie metody NearestCentroid jako algorytmu

```python
X, y = load_wine(return_X_y=True)
X2, X_test, y2, y_test = train_test_split(X, y,
                                                    test_size=0.10,
                                                    random_state=RANDOM_STATE)
X_train, X_valid, y_train, y_valid = train_test_split(X2, y2,
                                                    test_size=0.80,
                                                    random_state=RANDOM_STATE)


from sklearn.neighbors.nearest_centroid import NearestCentroid
import numpy as np

clf = NearestCentroid()
clf.fit(X_train, y_train)

clf.score(X_train, y_train)
clf.score(X_valid,y_valid)
```

Zmiana metody obliczeń na RandomForest.

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=150, max_depth=4,
                             random_state=RANDOM_STATE)
clf.fit(X_train, y_train)

print(clf.score(X_valid, y_valid))
```

Metoda RandomForest daje lepsze wyniki niż NearestCentroid.
